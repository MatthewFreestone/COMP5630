```python
import numpy as np
import matplotlib.pyplot as plt
```

## Part A

```python
samples = np.random.normal(loc=(0,0), size=(1000,2))
```

```python
x, y = samples.T
plt.scatter(x,y)
```

## Part B

```python
samples = np.random.normal(loc=(-1,1), size=(1000,2))
```

```python
x, y = samples.T
plt.scatter(x,y)
```

## Part C

```python
variance = 2
standard_dev = np.sqrt(variance)
samples = np.random.normal(loc=0, scale=[standard_dev, standard_dev],size=(1000,2))
```

```python
x, y = samples.T
plt.scatter(x,y)
```

## Part D

```python
covar_matrix = np.array([[1,0.5],[0.5,1]])
print(covar_matrix)
samples = np.random.normal(loc=0, scale=covar_matrix,size=(1000,2))
```

```python
x, y = samples.T
plt.scatter(x,y)
```
